```python
# Ch. 17: Interpolation techniques
# Sections: 17.1, 17.2, 17.3
# 17.1: Interpolation Statement
# Interpolation does not require a user to have underlying models for data. Instead the user assumes they have reliable data points
# that can be used to estimate a function for every point. Interpolation methods are used to determine function values.

# 17.1: interpolation
# estimate function that goes through data points

# P1: interpolation example
# Find the nearest interpolation at x=3.5 based on the data x = [2, 4, 6], y = [4, 6, 0].

# import array processing package 'numpy'
import numpy as np
# import plotting library with MATLAB-like interface
import matplotlib.pyplot as plt
# create matrices of x,y data given in PS
x = [2, 4, 6]
y = [4, 6, 0]
# set estimated x value data point
x_new = 3.5
# find new y data point using interpolation using x_new,x, and y as parameters
y_new = np.interp(x_new, x, y)
# print new y value
print("interpolated value =", y_new)

# plot function
plt.plot(x, y, 'o-')
plt.title('Interpolation at x = 3.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
# 17.2: Linear interpolation

# P2: Linear Interpolation Example 1
# An estimated data point is assumed to lie on the line joining the nearest points to the left and right.
# Find the linear interpolation at x=3.5 based on the data x = [2, 4, 6], y = [4, 6, 0].

# 1D function interpolation
from scipy.interpolate import interp1d
# import plotting library with MATLAB-like interface
import matplotlib.pyplot as plt

# seaborn is a statistical library used for plotting that adds visuale effects to plots to help with analyses. 
plt.style.use('seaborn-poster')
# create x,y matrices    
x = [2, 4, 6]
y = [4, 6, 0]

# estimate function using interpolation with x,y as parameters
f = interp1d(x, y)
# find y value by evaluation function at x value given in PS
y_hat = f(3.5)
# print yhat
print("interpolated value =",y_hat)

plt.plot(x, y, '-ob')
plt.plot(3.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 3.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

```python
# P3: Linear Interpolation Example 2
# An estimated data point is assumed to lie on the line joining the nearest points to the left and right.
# Find the linear interpolation at x=4.5 based on the data x = [2, 5, 6], y = [6, 3, 0].

# 1D function interpolation
from scipy.interpolate import interp1d
# import plotting library with MATLAB-like interface
import matplotlib.pyplot as plt

# seaborn is a statistical library used for plotting that adds visuale effects to plots to help with analyses. 
plt.style.use('seaborn-poster')
# create x,y matrices    
x = [2, 5, 6]
y = [6, 3, 0]

# estimate function using interpolation with x,y as parameters
f = interp1d(x, y)
# find y value by evaluation function at x value given in PS
y_hat = f(4.5)
# print yhat
print("interpolated value =", y_hat)

plt.plot(x, y, '-ob')
plt.plot(4.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 4.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 17.3: Cubic Spline Interpolation
# In cubic spline interpolation, the interpolating function is a set of piecewise cubic functions.

# P4: Cubic Spline Interpolation Example
# Use CubicSpline to plot the cubic spline interpolation of the data set x = [2, 5, 8] and y = [6, 8, 3] for 0≤x≤9.

# import CubicSpline interpolator
from scipy.interpolate import CubicSpline
# import array processing package 'numpy'
import numpy as np
# import plotting library with MATLAB-like interface
import matplotlib.pyplot as plt
     
# create x,y matrices
x = [2, 5, 8]
y = [6, 8, 2]

# estimate function f using CubicSpline and x,y,bc_type= ' ' as parameters. 
f = CubicSpline(x, y, bc_type = 'natural')
# create new x array
x_new = np.linspace(0, 9, 100)
# evaluate y at x_new
y_new = f(x_new)
     
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
     
```
