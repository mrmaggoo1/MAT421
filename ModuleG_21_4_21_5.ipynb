```python
# Sections 21.4, 21.5

# 21.4: Simpson's Rule

# P1: Use Simpson's Rule to approx. the integral of cos(x) from the interval [0,pi]
import numpy as np

a = 0
b = np.pi
n = 10
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_simp = (h/3) * (f[0] + 2*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 2 - I_simp

print(I_simp)
print(err_simp)
```

```python
# 21.5: Computing integrals in Python

# P2: Use trapz function to approx the integral of cos(x) from [0,pi] using n = 10.
import numpy as np
from scipy.integrate import trapz

a = 0
b = np.pi
n = 10
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_trapz = trapz(f,x)
I_trap = (h/2)*(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print(I_trapz)
print(I_trap)
```

```python
# P3: Use cumtrapz to approx. the cummulative integral for fx = cos(x) [0,pi] with h = 0.01
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0.1, np.pi, 0.01)
F_exact = np.sin(x)
F_approx = cumtrapz(np.cos(x), x)

plt.figure(figsize = (14,10))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} sin(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```

```python
# P4: Use integrate.quad to approx. cos(x) from [0,pi]. Compare to 0
from scipy.integrate import quad 

I_quad, est_err_quad = \
          quad(np.cos, 0, np.pi)
print(I_quad)
err_quad = 2 - I_quad
print(est_err_quad, err_quad)
```
